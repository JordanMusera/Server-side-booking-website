```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
import torch.nn.functional as F
class Model(nn.Module):
    def __init__(self, in_features=4, h1=8, h2=8, out_features=3):
        super().__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1, h2)
        self.out = nn.Linear(h2, out_features)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x
```

```python
torch.manual_seed(41)

#create an instance of the Model
model = Model()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
my_df = pd.read_csv(url)
my_df
```

```python
#change last column from string to integers
my_df['species'] = my_df['species'].replace('setosa',0.0)
my_df['species'] = my_df['species'].replace('versicolor',1.0)
my_df['species'] = my_df['species'].replace('virginica',2.0)

my_df
```

```python
#Train test split x,y
x = my_df.drop('species',axis=1)
y = my_df['species']

#convert these to numpy array
x = x.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
torch.manual_seed(41)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)

X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
#Set the creterion of the model tp measure the errors
criterion = nn.CrossEntropyLoss()
#Choose adam optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
epochs = 100
losses = []

for i in range(epochs):
    y_pred = model.forward(X_train)  # Get predicted result
    # Compute the loss
    loss = criterion(y_pred, y_train)

    # Store the loss
    losses.append(loss.detach().numpy())

    # Print loss every 10 epochs
    if i % 10 == 0:
        print(f'epoch {i} loss {loss}')

     # Backpropagation
    optimizer.zero_grad()  # Clear previous gradients
    loss.backward()  # Compute new gradients
    optimizer.step()  # Update weights
```

```python
plt.plot(range(epochs),losses)
plt.ylabel('Losses/errors')
plt.xlabel('epochs')
```

```python
new_iris = torch.tensor([[1.2, 2.0, 2.2, 0.2]])  # Ensure correct shape

with torch.no_grad():
    output = model(new_iris)  # Get model output
    predicted_class = torch.argmax(output, dim=1).item()  # Get class index

    # Map index to flower name
    if predicted_class == 0:
        x = 'Setosa'
    elif predicted_class == 1:
        x = 'Versicolor'
    else:
        x = 'Virginica'

    print(f'Predicted class: {x}')

```
